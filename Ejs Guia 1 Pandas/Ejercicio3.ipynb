3-Dado un dataframe, generar los siguientes dataframes
* Promedio por producto
* Mínimo, Máximo y Promedio de ventas por producto con formato 1
* Mínimo, Máximo y Promedio de ventas por producto con formato 2
* Mínimo, Máximo y Promedio de ventas por producto con formato 3
* Qué productos son provistos por los distintos fabricantes, indicando True/False

```python
import pandas as pd
```

```python
productos_df = pd.read_csv('Datasets/productos.csv')
productos_df.head()
```

```python
#Renombo las columnas para que quede como el enunciado(ver pdf)
columnas = {'Codigo_producto':'Código producto','Codigo_fabricante':'Código fabricante','ventas_mensuales':'Ventas Mensuales'}
productos_df.rename(columns = columnas,inplace=True)
```

**Promedio por producto**


```python
productos_agrupados = productos_df.groupby(by='Código producto')
promedio_ventas_df = round(productos_agrupados.agg({'Ventas Mensuales':'mean'}),0)

#Acomodo el nombre y lo muestro
nombre_nuevo={'Ventas Mensuales':'Promedio de ventas'}
promedio_ventas_df.rename(columns=nombre_nuevo,inplace=True)
promedio_ventas_df.head()
```

**Mínimo, Máximo y Promedio de ventas por producto con formato 1**

```python
round(productos_agrupados.agg({'Ventas Mensuales':['min','max','mean']}),0).stack().head(6) # Muestro solo las primeras 6 filas
```

**Mínimo, Máximo y Promedio de ventas por producto con formato 2**

```python
min_max_mean_ventas_df = round(productos_agrupados.agg({'Ventas Mensuales':['min','max','mean']}),0)

#Cambio los nombres a las columnas
columnas_acomodadas = {'min':'Minimo ventas mensual','max':'Maximo ventas mensual','mean':'Promedio ventas mensual'}
min_max_mean_ventas_df.rename(columns = columnas_acomodadas,inplace = True)

#Las acomodo y muestro como queda
min_max_mean_ventas_df.columns = min_max_mean_ventas_df.columns.get_level_values(1)
min_max_mean_ventas_df
```

**Mínimo, Máximo y Promedio de ventas por producto con formato 3**

```python
min_max_mean_ventas_df.T
```

**Qué productos son provistos por los distintos fabricantes, indicando True/False**

```python
productos = productos_df['Código producto'].unique() #Obtengo los valores unicos para los fabricantes y productos
fabricantes=productos_df['Código fabricante'].unique()

disponibilidad_fab_df = pd.DataFrame(index = productos , columns = fabricantes)
```

```python
fabricantes_agrupados = productos_df.groupby(by='Código fabricante')

# Mejor forma de hacer esto, algo nativo de pandas?
for fabricante in disponibilidad_fab_df.columns:
    for producto in disponibilidad_fab_df.index:
        productos_del_fabricante = fabricantes_agrupados.get_group(fabricante)['Código producto'].values
        disponibilidad_fab_df.at[producto,fabricante] = producto in productos_del_fabricante
        
disponibilidad_fab_df
```
