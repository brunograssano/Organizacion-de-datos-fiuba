4- Un sitio de Ebooks tiene información sobre los reviews que los usuarios hacen de sus libros
en un DataFrame con formato (user_id, book_id, rating, timestamp). Por otro lado tenemos
información en otro DataFrame que bajamos de GoodReads: (book_id, book_name,
avg_rating). Podemos suponer que los Ids de los libros son compatibles. Se pide usar Python
Pandas para:
* Obtener un DataFrame que indique el TOP5 de Ebooks en el sitio de Ebooks. (Para este punto se puede ignorar el segundo DataFrame).
* Obtener un DataFrame que indique qué libros tienen una diferencia de rating promedio mayor al 20% entre el sitio de Ebooks y GoodReads.

*Datasets generados por https://www.generatedata.com/*

*La columna de book_name va a tener nombres de quimicos por el generador :/*

```python
import pandas as pd
```

```python
#Cargo el Dataframe correspondiente al sitio de los Ebooks
ebooks_df = pd.read_csv('Datasets/SitioEbooks.csv')
ebooks_df.head()
```

```python
#Cargo el Dataframe correspondiente al sitio de GoodReads
good_reads_df = pd.read_csv('Datasets/GoodReads.csv')
good_reads_df.head()
```

**Obtener un DataFrame que indique el TOP5 de Ebooks en el sitio de Ebooks. (Para este punto se puede ignorar el segundo DataFrame).**

```python
mean_rating_ebook_df = ebooks_df.groupby(by='book_id').agg({'rating':'mean'})
mean_rating_ebook_df.nlargest(5,'rating')
```

**Obtener un DataFrame que indique qué libros tienen una diferencia de rating promedio mayor al 20% entre el sitio de Ebooks y GoodReads.**

```python
good_reads_df.set_index('book_id',inplace=True)
```

```python
good_reads_df.sort_index(ascending=True,inplace=True)
```

```python
#Agrego los libros que tienen el promedio de rating del sitio de Ebook al de GoodReads
good_reads_df['avg_rating_ebooks'] = mean_rating_ebook_df['rating']
```

```python
good_reads_df = good_reads_df.rename(columns = {'avg_rating':'avg_rating_goodreads'})
good_reads_df.head()
```

```python
good_reads_ebook_diff_df = good_reads_df.loc[(good_reads_df['avg_rating_ebooks']/good_reads_df['avg_rating_goodreads'])>0.2,:]
good_reads_ebook_diff_df
```
